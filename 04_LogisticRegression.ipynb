# Logistic Regression
- Logistic Regression 에 대한 자세한 내용을 설명하기 보다는 Tensorflow 2.0 을 익히기 위한 Notebook 입니다.

- Data 출처: [Kaggle Piama Indians Diabates Database](https://www.kaggle.com/uciml/pima-indians-diabetes-database)

```python
import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam

import numpy as np
import pandas as pd

print(tf.__version__)
```

```python
# Data Directory
DATA_DIR = './data/diabetes.csv'

# Loading Data
dataframe = pd.read_csv(DATA_DIR)
dataframe.head()
```

```python
# Data Generation
X_data = dataframe.loc[:, 'Pregnancies':'Age'].to_numpy()
Y_data = dataframe.loc[:, 'Outcome'].to_numpy()

print("X_data shape: {}".format(X_data.shape))
print("Y_data shape: {}".format(Y_data.shape))
```

```python
# Modeling
model = Sequential()
model.add(Dense(1, input_shape=(X_data.shape[1],), 
                activation='sigmoid'))

# Compiling
model.compile(optimizer=Adam(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])

# Confirmation
model.summary()
```

```python
# Training
# 데이터가 어느정도 충분히 있으므로, Validation Split을 이용하자
trained_model = model.fit(X_data, Y_data, epochs=1000, validation_split=0.2)
```

```python
# Evaluation
model.evaluate(X_data, Y_data)
```
